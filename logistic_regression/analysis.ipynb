## Arcene Dataset

#### SGD Testing, lr = 0.1, 0.01, 0.005, 0.001

```python
SGD_01_file_path = "arcene/GD/lr-0.1/200/no_batching/2024-09-30-16:03:48/results.csv"
SGD_001_file_path = "arcene/GD/lr-0.01/200/no_batching/2024-09-30-15:08:21/results.csv"
SGD_0005_file_path = "arcene/GD/lr-0.005/200/no_batching/2024-09-30-16:05:07/results.csv"
SGD_0001_file_path = "arcene/GD/lr-0.001/200/no_batching/2024-09-30-16:05:18/results.csv"

# Read CSV files
SGD_01_stats = pd.read_csv(SGD_01_file_path)
SGD_001_stats = pd.read_csv(SGD_001_file_path)
SGD_0005_stats = pd.read_csv(SGD_0005_file_path)
SGD_0001_stats = pd.read_csv(SGD_0001_file_path)
```

```python
# Extract data from DataFrames
epochs = SGD_01_stats['Epoch']  # Assuming all have the same epochs

# Extract data for various datasets
SGD_01_test_loss = SGD_01_stats['Test Loss']
SGD_001_test_loss = SGD_001_stats['Test Loss']
SGD_0005_test_loss = SGD_0005_stats['Test Loss']
SGD_0001_test_loss = SGD_0001_stats['Test Loss']

SGD_01_test_accuracy = SGD_01_stats['Test Accuracy']
SGD_001_test_accuracy = SGD_001_stats['Test Accuracy']
SGD_0005_test_accuracy = SGD_0005_stats['Test Accuracy']
SGD_0001_test_accuracy = SGD_0001_stats['Test Accuracy']

SGD_01_training_loss = SGD_01_stats['Training Loss']
SGD_001_training_loss = SGD_001_stats['Training Loss']
SGD_0005_training_loss = SGD_0005_stats['Training Loss']
SGD_0001_training_loss = SGD_0001_stats['Training Loss']

# Calculate total training time (expanding sum) for each dataset
SGD_01_total_training_time = SGD_01_stats['Training Time (s)'].expanding().sum()
SGD_001_total_training_time = SGD_001_stats['Training Time (s)'].expanding().sum()
SGD_0005_total_training_time = SGD_0005_stats['Training Time (s)'].expanding().sum()
SGD_0001_total_training_time = SGD_0001_stats['Training Time (s)'].expanding().sum()

# Create subplots in a 2x3 grid
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Plot 1: Test Loss vs Epochs
axes[0, 0].plot(epochs, SGD_01_test_loss, label='SGD, lr=0.1', color='red')
axes[0, 0].plot(epochs, SGD_001_test_loss, label='SGD, lr=0.01', color='blue')
axes[0, 0].plot(epochs, SGD_0005_test_loss, label='SGD, lr=0.005', color='green')
axes[0, 0].plot(epochs, SGD_0001_test_loss, label='SGD, lr=0.001', color='orange')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Test Loss (Cross-Entropy)')
axes[0, 0].set_title('Test Loss vs Epochs')
axes[0, 0].legend()

# Plot 2: Test Accuracy vs Epochs
axes[0, 1].plot(epochs, SGD_01_test_accuracy, label='SGD, lr=0.1', color='red')
axes[0, 1].plot(epochs, SGD_001_test_accuracy, label='SGD, lr=0.01', color='blue')
axes[0, 1].plot(epochs, SGD_0005_test_accuracy, label='SGD, lr=0.005', color='green')
axes[0, 1].plot(epochs, SGD_0001_test_accuracy, label='SGD, lr=0.001', color='orange')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Test Accuracy')
axes[0, 1].set_title('Test Accuracy vs Epochs')
axes[0, 1].legend()

# Plot 3: Training Loss vs Epochs
axes[0, 2].plot(epochs, SGD_01_training_loss, label='SGD, lr=0.1', color='red')
axes[0, 2].plot(epochs, SGD_001_training_loss, label='SGD, lr=0.01', color='blue')
axes[0, 2].plot(epochs, SGD_0005_training_loss, label='SGD, lr=0.005', color='green')
axes[0, 2].plot(epochs, SGD_0001_training_loss, label='SGD, lr=0.001', color='orange')
axes[0, 2].set_xlabel('Epoch')
axes[0, 2].set_ylabel('Training Loss (Cross-Entropy)')
axes[0, 2].set_title('Training Loss vs Epochs')
axes[0, 2].legend()

# Plot 4: Test Loss vs Training Time
axes[1, 0].plot(SGD_01_total_training_time, SGD_01_test_loss, label='SGD, lr=0.1', color='red')
axes[1, 0].plot(SGD_001_total_training_time, SGD_001_test_loss, label='SGD, lr=0.01', color='blue')
axes[1, 0].plot(SGD_0005_total_training_time, SGD_0005_test_loss, label='SGD, lr=0.005', color='green')
axes[1, 0].plot(SGD_0001_total_training_time, SGD_0001_test_loss, label='SGD, lr=0.001', color='orange')
axes[1, 0].set_xlabel('Training Time (s)')
axes[1, 0].set_ylabel('Test Loss (Cross-Entropy)')
axes[1, 0].set_title('Test Loss vs Training Time')
axes[1, 0].legend()

# Plot 5: Test Accuracy vs Training Time
axes[1, 1].plot(SGD_01_total_training_time, SGD_01_test_accuracy, label='SGD, lr=0.1', color='red')
axes[1, 1].plot(SGD_001_total_training_time, SGD_001_test_accuracy, label='SGD, lr=0.01', color='blue')
axes[1, 1].plot(SGD_0005_total_training_time, SGD_0005_test_accuracy, label='SGD, lr=0.005', color='green')
axes[1, 1].plot(SGD_0001_total_training_time, SGD_0001_test_accuracy, label='SGD, lr=0.001', color='orange')
axes[1, 1].set_xlabel('Training Time (s)')
axes[1, 1].set_ylabel('Test Accuracy')
axes[1, 1].set_title('Test Accuracy vs Training Time')
axes[1, 1].legend()

# Plot 6: Training Loss vs Training Time
axes[1, 2].plot(SGD_01_total_training_time, SGD_01_training_loss, label='SGD, lr=0.1', color='red')
axes[1, 2].plot(SGD_001_total_training_time, SGD_001_training_loss, label='SGD, lr=0.01', color='blue')
axes[1, 2].plot(SGD_0005_total_training_time, SGD_0005_training_loss, label='SGD, lr=0.005', color='green')
axes[1, 2].plot(SGD_0001_total_training_time, SGD_0001_training_loss, label='SGD, lr=0.001', color='orange')
axes[1, 2].set_xlabel('Training Time (s)')
axes[1, 2].set_ylabel('Training Loss (Cross-Entropy)')
axes[1, 2].set_title('Training Loss vs Training Time')
axes[1, 2].legend()

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()

```
