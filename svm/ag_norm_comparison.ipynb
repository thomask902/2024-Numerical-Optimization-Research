```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("generated/hinge/n_2000_m_1000/AG/no-lr/200/no-batching/2024-12-17-12:24:23/results.csv")
df.head(50)
```

```python
# New gradient norms for overlay
epochs_list = df["Epoch"]
x_md_k_list = df['x_md_k gradient norm']
x_ag_k_list = df['x_ag_k gradient norm']
x_k_list = df['x_k gradient norm']

# Force specific width and height
fig, ax = plt.subplots(figsize=(15, 5))  # 15 inches wide, 5 inches tall

ax.plot(epochs_list, x_md_k_list, label='x_md_k', color='red')
ax.plot(epochs_list, x_ag_k_list, label='x_ag_k', color='green')
ax.plot(epochs_list, x_k_list, label='x_k', color='blue')

ax.set_xlabel('Epoch')
ax.set_ylabel('Gradient Norm')
ax.set_title('Overlay of x_md_k, x_ag_k, and x_k Gradient Norms over Epochs')
ax.legend()

plt.tight_layout()
plt.show()
```
